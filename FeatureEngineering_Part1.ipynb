```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

### (1) Are the exam scores (math, reading, writing) normally distributed? If not, what can we do to have a normal distribution?
(Sınav puanları (matematik, okuma, yazma) normal dağılmış mı? Değilse, normal dağılıma sahip olmak için ne yapabiliriz?)

```python
df.dropna(inplace=True)
```

```python
from scipy.stats import jarque_bera
from scipy.stats import normaltest

pd.options.display.float_format = '{:.5f}'.format

variables = ['writing score', 'reading score' , 'math score']
distribution_tests = pd.DataFrame(columns=['variable', 'jarque_bera_stats', 'jarque_bera_p_value'])

for variable in variables:
    jb_stats = jarque_bera(df[variable])
    distribution_tests = distribution_tests.append({"variable": variable,
                                                "jarque_bera_stats" : jb_stats[0] ,
                                                "jarque_bera_p_value" : jb_stats[1] 
                                               }, ignore_index=True)
distribution_tests
```

p value'yu değerlendirdiğimizde değerler alfa'dan küçük çıkmıştır. Bu nedenle değerler normal dağılmamıştır diyebiliriz.

Normal dağılmamış değerler için log alma, winsorize uygulama, karekökünü alma, tersini alma gibi yöntemler uygulanabilir.

### (4) Does the test preparation course have an impact on the target variable you created?
#### (Teste hazırlık kursunun oluşturduğunuz hedef değişken üzerinde etkisi var mı?)

```python
df.groupby("test preparation course").mean().T
```

```python
import scipy.stats as stats
testPrep = df["test preparation course"].dropna().unique()
pd.options.display.float_format = '{:.15f}'.format
for var in ["all_scores_mean"]:
    comparison = pd.DataFrame(columns=['group_1', 'group_2','statistic', 'p_value'])
    print("For the sake of comparison {}".format(var),end='')
    for i in range(0, len(testPrep)):
        for j in range(i+1, len(testPrep)):
            ttest = stats.ttest_ind(df[df["test preparation course"]==testPrep[i]][var], 
                                df[df["test preparation course"]==testPrep[j]][var])
            group_1 = testPrep[i]
            group_2 = testPrep[j]
            statistic = ttest[0]
            p_value = ttest[1]
            
            comparison = comparison.append({"group_1" : group_1 ,
                                                  "group_2" : group_2 ,
                                                  "statistic": statistic , 
                                                  "p_value" : p_value}, ignore_index=True)
    display(comparison)
```

Aralarında bir ilişki gözlemlenememiştir.
